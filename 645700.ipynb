EXPLICACION CUATRO: Para nosotros poder implementar una grafica, debemos llamar a la función de matplotlib inline, que nos concede mostrar las graficas en la pantalla de manera correcta al trabajar con Jupyter Notebook o bien algun entorno parecido; importamos la librería de matplotlib.pyplot abreviando esta como plt, para implementar el uso de los recursos gráficos. Definimos dos variables "y" y "x" a las cuales les damos los datos de Nivel de Obesidad y Peso respectivamente; a cada uno les daremos una condición que deberá englobar solo a la población femenina en los datos que se enseñarán en la gráfica, condicionando esto a que solo se tome a las personas registradas como "Female" en los datos de Sexo. 
Por ultimo con nuestra funcion scatter, hacemos que cada dato se enseñe como un punto en la gráfica, mostrando su relación con los datos puestos en cada eje, dentro de scatter definí un alpha como 0.023, que solo es para darle una transparencia al punto y poder ver donde se aglomeran mas estos.
La función show la usé para mostrar la gráfica en la consola.

```python
%matplotlib inline
import matplotlib.pyplot as plt
y= datos.NivelDeObesidad[datos.Sexo == "Female"]
x= datos.Peso[datos.Sexo == "Female"]

plt.scatter(x, y, alpha=0.023)
plt.show()
```
